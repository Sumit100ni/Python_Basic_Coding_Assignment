# 3.	Write a Python program to find the area of a triangle?

```python
h=int(input('Enter the Height of the Triangle: '))
b=int(input('Enter the Base of the Triangle: '))
area=(1/2)*h*b
print('The Area of the Triangle is: ',area)
```
